10. What is difference between the list methods append() and insert()?

Ans: append() adds the element at the end of the list whereas insert() adds the element at the specified index.
append() takes only one argument i.e. object whereas insert() takes two arguments i.e. index and object

Below we can see example of append() and insert()

```python
l1 = ["a","b","c"]
l1.append("k")
l1
# this is append()
```

```python
l1 = ["a","b","c"]
l1.insert(1,"k")
l1
# this is insert
```

17. How do you distinguish between copy.copy() and copy.deepcopy()?

Ans: when we use copy.copy(), it is a shallow copy. In shallow copy if we change an element in the new list, it will be changed in  both the new list and old list.(only in nested list)

When we use copy.deepcopy(), it is a deep copy. In deep copy if we change an element in the new list, it will be changed only in the new list but the old list will not be affected.

We can see the above changes only in case of nested list. However in case of normal list, output of shallow copy & deep copy are same.

in normal list shallow copy is equal to deep copy

```python
lst1 = [1,2,3,4,5]
```

```python
# shallow copy
lst2 = lst1.copy()
```

```python
lst2
```

```python
lst2[1]= 100
lst2
```

```python
lst1
```

```python
lst1 = [[1,2,3,4],[5,6,7,8]]
lst2 = lst1.copy()
# shallow copy of nested list
```

```python
lst2[1][0] = 100
```

```python
lst2
```

```python
lst1
```

We can see that in case of shallow copy, the number changed in lst2 is also changed in lst1

```python
import copy
lst1 = [[1,2,3,4],[5,6,7,8]]
lst2 = copy.deepcopy(lst1)
#deep copy of nested list
```

```python
lst2[1][0] = 100
```

```python
lst2
```

```python
lst1
```

We can see that in case of deepow copy, the number changed in lst2 is not changed in lst1
